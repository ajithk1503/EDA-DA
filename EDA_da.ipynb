```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/telco%20customer%20churn.csv")
data
```

```python
# dimmensions
print("dimensions of the data set: ",data.shape)
```

```python
# summary
print("summary of the data set: ")
print(data.info())
```

## **Data Handling and Cleaning**

```python
# finding missing values
print(data.isnull().sum())
```

```python
# drop duplicates
data = data.drop_duplicates()
```

```python
data.tenure = data.tenure.replace(np.nan,0)
data.tenure = pd.to_numeric(data.tenure, errors='coerce')
data.tenure = data.tenure.fillna(data.tenure.mean())
```

```python
# filling missing values
data.SeniorCitizen=data.SeniorCitizen.fillna(0)
data.Partner=data.Partner.fillna('No')
data.tenure = data.tenure.replace(np.nan,0)
data.tenure = pd.to_numeric(data.tenure, errors='coerce')
data.tenure = data.tenure.fillna(data.tenure.mean())
data.MonthlyCharges = data.MonthlyCharges.fillna(data.MonthlyCharges.mean())
data.isnull().sum()
```

```python
# removing a row as the result [Churn] is unknown
data=data.dropna()
```

```python
data.isnull().sum()
```
