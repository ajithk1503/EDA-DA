## **Data Handling and Cleaning**

```python
# finding missing values
print(data.isnull().sum())
```

```python
# drop duplicates
data = data.drop_duplicates()
```

```python
data.tenure = data.tenure.replace(np.nan,0)
data.tenure = pd.to_numeric(data.tenure, errors='coerce')
data.tenure = data.tenure.fillna(data.tenure.mean())
```

```python
# filling missing values
data.SeniorCitizen=data.SeniorCitizen.fillna(0)
data.Partner=data.Partner.fillna('No')
data.tenure = data.tenure.replace(np.nan,0)
data.tenure = pd.to_numeric(data.tenure, errors='coerce')
data.tenure = data.tenure.fillna(data.tenure.mean())
data.MonthlyCharges = data.MonthlyCharges.fillna(data.MonthlyCharges.mean())
data.isnull().sum()
```

```python
# removing a row as the result [Churn] is unknown
data=data.dropna()
```

## Measures of Dispersion

```python
from scipy import stats
```

```python
# range
# range of tenure
data_range = np.ptp(data.tenure)
print("Range:", data_range)
# range of monthly charges
data_range = np.ptp(data.MonthlyCharges)
print("Range:", data_range)
# range of total charges
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
data_range = np.ptp(data_cleaned['TotalCharges'])
print("Range:", data_range)
```

```python
# Interquartile Range (IQR)
# IQR for 'tenure'
iqr_tenure = np.percentile(data.tenure, 75) - np.percentile(data.tenure, 25)
print("IQR (tenure):", iqr_tenure)
# IQR for 'MonthlyCharges'
iqr_monthly_charges = np.percentile(data.MonthlyCharges, 75) - np.percentile(data.MonthlyCharges, 25)
print("IQR (MonthlyCharges):", iqr_monthly_charges)
# IQR for 'TotalCharges'
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
iqr_total_charges = np.percentile(data_cleaned['TotalCharges'], 75) - np.percentile(data_cleaned['TotalCharges'], 25)
print("IQR (TotalCharges):", iqr_total_charges)
```

```python
# Interdecile Range
# Interdecile Range for 'tenure'
interdecile_range_tenure = np.percentile(data.tenure, 90) - np.percentile(data.tenure, 10)
print("Interdecile Range (tenure):", interdecile_range_tenure)

# Interdecile Range for 'MonthlyCharges'
interdecile_range_monthly_charges = np.percentile(data.MonthlyCharges, 90) - np.percentile(data.MonthlyCharges, 10)
print("Interdecile Range (MonthlyCharges):", interdecile_range_monthly_charges)

# Interdecile Range for 'TotalCharges'
interdecile_range_total_charges = np.percentile(data_cleaned['TotalCharges'], 90) - np.percentile(data_cleaned['TotalCharges'], 10)
print("Interdecile Range (TotalCharges):", interdecile_range_total_charges)
```

```python
# Mean Deviation
# Mean Deviation for 'tenure'
mean_deviation_tenure = np.mean(np.abs(data.tenure - np.mean(data.tenure)))
print("Mean Deviation (tenure):", mean_deviation_tenure)
# Mean Deviation for 'MonthlyCharges'
mean_deviation_monthly_charges = np.mean(np.abs(data.MonthlyCharges - np.mean(data.MonthlyCharges)))
print("Mean Deviation (MonthlyCharges):", mean_deviation_monthly_charges)
# Mean Deviation for 'TotalCharges'
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
mean_deviation_total_charges = np.mean(np.abs(data_cleaned['TotalCharges'] - np.mean(data_cleaned['TotalCharges'])))
print("Mean Deviation (TotalCharges):", mean_deviation_total_charges)
```

```python
# Standard Deviation
# Standard Deviation for 'tenure'
std_dev_tenure = np.std(data.tenure, ddof=1)
print("Standard Deviation (tenure):", std_dev_tenure)
# Standard Deviation for 'MonthlyCharges'
std_dev_monthly_charges = np.std(data.MonthlyCharges, ddof=1)
print("Standard Deviation (MonthlyCharges):", std_dev_monthly_charges)
# Standard Deviation for 'TotalCharges'
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
std_dev_total_charges = np.std(data_cleaned['TotalCharges'], ddof=1)
print("Standard Deviation (TotalCharges):", std_dev_total_charges)
```

```python
# Skewness
# Skewness for 'tenure'
skewness_tenure = stats.skew(data.tenure)
print("Skewness (tenure):", skewness_tenure)
# Skewness for 'MonthlyCharges'
skewness_monthly_charges = stats.skew(data.MonthlyCharges)
print("Skewness (MonthlyCharges):", skewness_monthly_charges)
# Skewness for 'TotalCharges'
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
skewness_total_charges = stats.skew(data_cleaned['TotalCharges'])
print("Skewness (TotalCharges):", skewness_total_charges)
```

```python
# Kurtosis
# Kurtosis for 'tenure'
kurtosis_tenure = stats.kurtosis(data.tenure)
print("Kurtosis (tenure):", kurtosis_tenure)
# Kurtosis for 'MonthlyCharges'
kurtosis_monthly_charges = stats.kurtosis(data.MonthlyCharges)
print("Kurtosis (MonthlyCharges):", kurtosis_monthly_charges)
# Kurtosis for 'TotalCharges'
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data_cleaned = data.dropna(subset=['TotalCharges'])
kurtosis_total_charges = stats.kurtosis(data_cleaned['TotalCharges'])
print("Kurtosis (TotalCharges):", kurtosis_total_charges)
```

## 2-D Statistical data Analysis

```python
# Contingency Table

# Categorical-Categorical
contingency_table = pd.crosstab(data['gender'], data['Partner'])
print("Contingency Table (Categorical-Categorical):")
print(contingency_table)

# row profile
row_profile = contingency_table.div(contingency_table.sum(axis=1), axis=0)
print("\nRow Profile Dataset:")
print(row_profile)

# column profile
column_profile = contingency_table.div(contingency_table.sum(axis=0), axis=1)
print("\nColumn Profile Dataset:")
print(column_profile)

# relative frequency dataset
relative_frequency = contingency_table / contingency_table.sum().sum()
print("\nRelative Frequency Dataset:")
print(relative_frequency)

#chi squared dataset
from scipy.stats import chi2_contingency
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
chi_square_table = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print("\nChi-Square Dataset (Expected Frequencies):")
print(chi_square_table)
```

```python
# Numerical - Categorical
data['Tenure_gender'] = pd.cut(data['tenure'], bins=[0, 20, 40, 60], labels=['Low', 'Medium', 'High'])
contingency_num_cat = pd.crosstab(data['Tenure_gender'], data['gender'])
print("\nContingency Table (Numerical-Categorical):")
print(contingency_num_cat)
```

```python
# Numerical - Numerical
correlation_matrix = data[['MonthlyCharges', 'TotalCharges']].corr()
print("\nCorrelation Matrix (Numerical-Numerical):")
print(correlation_matrix)
```

```python
#  three-way contingency table
contingency_table_3way = pd.crosstab(index=[data['DeviceProtection'], data['StreamingMovies']], columns=data['StreamingTV'])
print("Three-Way Contingency Table:")
print(contingency_table_3way)
```
